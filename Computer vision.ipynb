<h1>obeject detection<h1/>

```python
cap=cv.VideoCapture(0)
while True:
    _,frame=cap.read()
    hsv=cv.cvtColor(frame,cv.COLOR_BGR2HSV)
    lower=np.array([20,100,100])
    upper=np.array([40,255,255])
    mask=cv.inRange(hsv,lower,upper)
    res=cv.bitwise_and(frame,frame,mask=mask)
    cv.imshow('frame',frame)
    cv.imshow('mask',mask)
    cv.imshow('res',res)
    k = cv.waitKey(5) & 0xFF
    if k == 27:
        break

cv.destroyAllWindows()
cap.release()
```

<h1>morphological transformation</h1>

```python
#eroision
img=cv.imread(r"/home/rohit/Downloads/WhatsApp Image 2024-07-19 at 5.36.23 PM.jpeg")
img_=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
kernal=np.ones((2,2),np.uint8)
ercode=cv.erode(img_,kernal,1)
while True:
    cv.imshow("image",img_)
    cv.imshow("dst",ercode)
    key = cv.waitKey(20) & 0xFF
    if key == 27:  
        break
cv.destroyAllWindows()
```

```python
# dilation
dil=cv.dilate(img_,kernal,10)
while True:
    cv.imshow("image",img_)
    cv.imshow("dst",dil)
    key = cv.waitKey(20) & 0xFF
    if key == 27:  
        break
cv.destroyAllWindows()
```

```python
# opening
open=cv.morphologyEx(img_,cv.MORPH_OPEN,kernal)
# close
close=cv.morphologyEx(img_,cv.MORPH_CLOSE,kernal)
while True:
    cv.imshow("image",img_)
    cv.imshow("close",close)
    cv.imshow("open",open)
    key = cv.waitKey(20) & 0xFF
    if key == 27:  
        break
cv.destroyAllWindows()
```

```python
# grediant
gr=cv.morphologyEx(img_,cv.MORPH_GRADIENT,kernal)
while True:
    cv.imshow("image",img_)
    cv.imshow("dst",gr)
    key = cv.waitKey(20) & 0xFF
    if key == 27:  
        break
cv.destroyAllWindows()
```
