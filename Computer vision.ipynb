<h1>saving video</h1>

```python
vid=cv.VideoCapture(0)
forcc=cv.VideoWriter_fourcc(*'XVID')
out=cv.VideoWriter("vid.avi",forcc,20.0,(640,480))
```

```python
while True:
    ret,frame=vid.read()
    if ret==False:
        print("cant")
    else:
        out.write(frame)
        cv.imshow("frame",frame)
        if cv.waitKey(1)&0XFF == ord("q"):
            break
vid.release()
out.release()
cv.destroyAllWindows()
```

<h1>draw</h1>

```python
import numpy as np
img=np.zeros((512,512,3))
cv.imshow("img",img)
cv.waitKey(0)
```

line

```python
img=cv.line(img,(0,0),(512,512),(155,0,0),5)
cv.imshow("img",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

rectangle

```python
img=cv.rectangle(img,(10,200),(500,500),(0,155,0),10)
cv.imshow("img",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

circle

```python
img=cv.circle(img,(200,200),50,(0,0,255),5)
cv.imshow("img",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

polyline

```python
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts.reshape((-1,1,2))
img=cv.polylines(img,[pts],True,(0,0,255))
cv.imshow("img",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

text

```python
font=cv.FONT_HERSHEY_SIMPLEX
img=cv.putText(img,"rohit maurya",(10,500),font,0.5,(255,255,255),2,cv.LINE_AA)
cv.imshow("img",img)
cv.waitKey(0)
cv.destroyAllWindows()
```
